```python
import spacy
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

def create_heatmap(smlrity, cmap = "YlGnBu"):
    df = pd.DataFrame(smlrity)
    df.columns = labels
    df.index = labels
    fig, ax = plt.subplots(figsize=(5,5))
    sns.heatmap(df, cmap=cmap)

headlines = [
'Investors unfazed by correction as crypto funds see $154 million inflows',
'Investors inmutable by corection as cripto fonds see $154 million unflows',
]

labels = [headline[:20] for headline in headlines]

model = spacy.load('es_core_news_md')
docs = [model(headline) for headline in headlines]
# print(docs[0].vector)

similarity = []
for i in range(len(docs)):
    row = []
    for j in range(len(docs)):
      row.append(docs[i].similarity(docs[j]))
    similarity.append(row)
create_heatmap(similarity)

similarity
```
